# 서울시 반려견 놀이터 지도 시각화

```python
import pandas as pd

df = pd.read_csv('D:\kimomin\DataAnalysis\data\서울시_반려견_놀이터.csv')
df
```

```python
df.index
```

```python
import folium
from folium.plugins import MarkerCluster 

# folium 개체 생성
center_la = df['Latitude'].mean()
center_long = df['Longitude'].mean()
map = folium.Map(location=[center_la,center_long], zoom_start=13)

marker = MarkerCluster().add_to(map)

for i in df.index:
    folium.Marker(location=[df.loc[i,'Latitude'], df.loc[i,'Longitude']],
                  popup=df.loc[i,'공원명'],
                  icon=folium.Icon(color='blue',icon='dog',prefix='fa')).add_to(marker)
    
map.save('map.html')
```
