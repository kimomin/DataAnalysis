```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
happy_life = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_삶의만족도.xlsx',engine='openpyxl')
happy_health = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_건강.xlsx',engine='openpyxl')
happy_safe = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_안전.xlsx',engine='openpyxl')
happy_environ = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_환경.xlsx',engine='openpyxl')
happy_econo = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_경제.xlsx',engine='openpyxl')
happy_edu = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_교육.xlsx',engine='openpyxl')
happy_relation = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_관계및사회참여.xlsx',engine='openpyxl')
happy_leisure = pd.read_excel(r'D:\kimomin\DataAnalysis\happy\대한민국행복지도_여가.xlsx',engine='openpyxl')
```

```python
life_merge = happy_life.groupby('시도')[['삶의 만족도']].mean()
life_merge.head()
```

```python
temp1 = happy_health.groupby('시도')[['평균']].mean()
temp2 = happy_safe.groupby('시도')[['평균']].mean()
temp3 = happy_environ.groupby('시도')[['평균']].mean()
temp4 = happy_econo.groupby('시도')[['평균']].mean()
temp5 = happy_edu.groupby('시도')[['평균']].mean()
temp6 = happy_relation.groupby('시도')[['평균']].mean()
temp7 = happy_leisure.groupby('시도')[['평균']].mean()
```

```python
life_merge = pd.merge(life_merge,temp1.rename(columns={'평균':'건강'}),on='시도')
life_merge = pd.merge(life_merge,temp2.rename(columns={'평균':'안전'}),on='시도')
life_merge = pd.merge(life_merge,temp3.rename(columns={'평균':'환경'}),on='시도')
life_merge = pd.merge(life_merge,temp4.rename(columns={'평균':'경제'}),on='시도')
life_merge = pd.merge(life_merge,temp5.rename(columns={'평균':'교육'}),on='시도')
life_merge = pd.merge(life_merge,temp6.rename(columns={'평균':'관계'}),on='시도')
life_merge = pd.merge(life_merge,temp7.rename(columns={'평균':'여기'}),on='시도')
```
