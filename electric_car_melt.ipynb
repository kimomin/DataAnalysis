```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\kimomin\DataAnalysis\data\한국전력공사_지역별 전기차 현황정보_20240731.csv')
df
```

```python
df_melt = df.melt(id_vars=['기준일'], var_name='지역', value_name='자동차수')
```

```python
# 파생변수, '년', '월'

df_melt[['년','월','일']] = df_melt['기준일'].str.split('-', expand=True)   #스플릿
df_melt
```

```python
df_melt['년도'] = df_melt['기준일'].str[:4]   # 슬라이싱
```

```python
df_melt.info()
```

```python
df_melt['년도'] = df_melt['년도'].astype(int)
```

# 지역별, 년도별 분석

```python
# df_melt.groupby(['지역','년'],as_index=False)[['자동차수']].mean()

# 지역별 년도별 자동차수 평균, pivot_table

year_tot = round(df_melt.pivot_table(index='지역',columns='년',values='자동차수',aggfunc='mean'),1) # round는 소수점 설정
year_tot
```

# 2024년 분기별 분석

```python
# 2024년 데이터 추출
df_melt_2024 = df_melt.query('년=="2024"')
df_melt_2024.info()
```

```python
# 월의 데이터 타입을 정수로 변환

df_melt_2024['월'] = df_melt_2024['월'].astype(int)
df_melt_2024.info()
```

# 사용자 정의 함수 분기 계산

```python
def get_quarter(month):
    if month >=1 and month <=3:
        return '1분기'
    elif 4 <= month <= 6:
        return '2분기'
    elif 7 <= month <= 9:
        return '3분기'
    elif 10 <= month <= 12:
        return '4분기'
    else:
        return '잘못된 값'
    
df_melt_2024['분기'] = df_melt_2024['월'].map(get_quarter)
```

```python
df_melt_2024.head()
```

```python
# lambda 이용한 분긱 계산
df_melt_2024['분기_2'] = df_melt_2024['월'].map(lambda month: '1분기' if  1 <= month <= 3
                                             else '2분기' if 4 <= month <= 6
                                             else '3분기' if 7 <= month <= 9
                                             else '4분기'if 10 <= month <= 12
                                             else '잘못된 값')
```

```python
df_melt_2024.head()
```

```python
# 지역별 분기별 자동차수 평균, pivot table

df_2024 = df_melt_2024.pivot_table(index='지역',columns='분기',values='자동차수',aggfunc='mean')
```

```python
df_2024.plot(kind='bar')
```
